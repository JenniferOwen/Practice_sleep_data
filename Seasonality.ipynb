```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 

import os
import urllib.request
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import datetime as dt

# First, convert your strings to instances of Python datetime.date:
import datetime as dt

# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
df = pd.read_csv("clean_sleepdata.csv") 

# Preview the first 5 lines of the loaded data 
df.head()
df['Date'] = pd.to_datetime(df['Date'])

df.info()
```

```python
# Create the plot space upon which to plot the data
fig, ax = plt.subplots(figsize=(20, 7))

# Add the x-axis and the y-axis to the plot
ax.plot(df.Date.values,
        df['Sleep quality'], '-',
        color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Sleep quality (%)",
       title="Sleep quality over time")

# Clean up the x axis dates (reviewed in lesson 4)
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(DateFormatter("%m/%Y"))
plt.xticks(rotation=45)


plt.show()
```

```python
# Create the plot space upon which to plot the data
fig, ax = plt.subplots(figsize=(20, 7))

# Add the x-axis and the y-axis to the plot
ax.plot(df.Date.values,
        df['Time in bed in minutes'], '-',
        color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Time in bed (in minutes)",
       title="Time in bed")

# Clean up the x axis dates (reviewed in lesson 4)
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(DateFormatter("%m/%Y"))
plt.xticks(rotation=45)


plt.show()
```
